```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
df_crash = sns.load_dataset('car_crashes')
df_crash.head(3)
```

<h2>Distribution Plots<h2>
<h3>These include Distribution Plot, Joint Plot, KDE Plot, Pair Plot etc.<h3>

```python
#We'll first discuss distribution plot. It is used to plot a distribution of one variable.
sns.displot(df_crash['not_distracted'])
```

```python
#Include the Kernel Density Estimation(KDE) line.
sns.displot(df_crash['not_distracted'],kde=True)
```

```python
#Alter the bin width.
sns.displot(df_crash['not_distracted'],bins=20)
```

```python
#We'll now look at Joint Plots. These are used to plot a distribution of two variables as a scatter plot.
sns.jointplot(x='speeding',y='alcohol',data=df_crash,kind='reg')
```

```python
sns.jointplot(x='speeding',y='alcohol',data=df_crash,kind='kde')
```

```python
sns.jointplot(x='speeding',y='alcohol',data=df_crash,kind='hex')
```

```python
#We'll now move to Pair Plots. It creates a matrix of plots showing relationship b/w all numeric variables.
#We'll first load 'tips' dataset.
df_tips = sns.load_dataset('tips')
df_tips.head(3)
```

```python
sns.pairplot(df_tips,hue='sex',palette='afmhot')
plt.gcf().set_size_inches(15,8) 
#Alter the size of the output plot using 'plt.gcf()' method. Used AFTER the plot statement.
```

<h2>Styling<h2>

```python
sns.set_style('darkgrid') #Dark background with grids.
sns.jointplot(x='speeding',y='alcohol',data=df_crash,kind='reg')
```

```python
sns.set_style('white') #White background. We also have 'whitegrid', 'dark' and 'ticks' to name a few popular options.
sns.set_context('talk',font_scale=1.5) #The 'talk/paper' option sets the style.
sns.jointplot(x='speeding',y='alcohol',data=df_crash,kind='reg')
plt.gcf().set_size_inches(13,8)
```

<h2>Bar Plots<h2>
<h3>Used when we have 1 categorical and 1 numeric variable.<h3>
<h3>The 'sns.barplot' function calculates a summary statistic for each category. By default it is mean.<h3>
<h3>Hence the height of the barplot depicts mean for that category.<h3>

```python
sns.barplot(data=df_tips,x='sex',y='total_bill',estimator=np.median)
plt.gcf().set_size_inches(10,6.5)
```

```python
#The 'countplot' is a special type of Bar Plot where the y-axis has counts for the categories of x-axis.
sns.countplot(x='sex',data=df_tips)
```

<h2>Box Plots<h2>

```python
sns.boxplot(data=df_tips,x='sex',y='total_bill')
```

```python
sns.boxplot(data=df_tips,x='day', y='total_bill',hue='sex')
plt.gcf().set_size_inches(14,8.5) #The 'hue' option generates boxplots for each level of 'sex' column.
plt.legend(loc='upper left',borderpad=0.1)
```

<h2>Violin Plots<h2>

```python
sns.violinplot(data=df_tips,x='day',y='total_bill',hue='sex')
plt.gcf().set_size_inches(15,8.5)
plt.legend(loc='upper left',borderpad=0.1)
```

```python
sns.violinplot(data=df_tips,x='day',y='total_bill',hue='sex',split=True)
plt.gcf().set_size_inches(15,8.5)
plt.legend(loc='upper left',borderpad=0.1)
```

<h2>Strip Plots<h2>

```python
sns.stripplot(data=df_tips,x='day',y='total_bill',jitter=True)
plt.gcf().set_size_inches(12,8.5)
```

```python
sns.stripplot(data=df_tips,x='day',y='total_bill',jitter=True,hue='sex')
plt.gcf().set_size_inches(12,8.5) #The 'jitter=True' sets the data points a bit apart rather than close to each other.
plt.legend(loc='upper left',borderpad=0.1)
```

```python
sns.stripplot(data=df_tips,x='day',y='total_bill',jitter=True,hue='sex',dodge=True)
plt.gcf().set_size_inches(12,8.5)
plt.legend(loc='upper left',borderpad=0.1) #The 'dodge=True' option sets the Male and Female data points apart.
```

<h2>Palettes<h2>

```python
sns.set_style('whitegrid')
sns.set_context('talk')
sns.barplot(data=df_tips,x='day',y='total_bill',palette='afmhot',hue='sex',ci=None)
plt.gcf().set_size_inches(12,8.5)
```

<h2>HeatMaps<h2>

```python
sns.set_context('talk',font_scale=0.8)
sns.set_style('dark')
cor_mx = df_crash.corr().round(2)
cor_mx #Create a correlation matrix b/w each numeric column of dataframe.
```

```python
sns.heatmap(cor_mx,annot=True,cmap='BuPu') 
```

```python
#Another method to plot heatmaps b/w two categorical variables and observe the frequency/heat along the pivot.
flights = sns.load_dataset('flights')
pivot = flights.pivot_table(index='month',columns='year', values='passengers')
pivot #Create a pivot table b/w two categorical variables.
```

```python
sns.set_context('talk',font_scale=0.5)
sns.heatmap(pivot,cmap='YlGn',linecolor='black',linewidth=1)
plt.gcf().set_size_inches(14,9)
#The 'cmap' option is used for color pattern in heat map.
#The 'linecolor' and 'linewidth' options decide the color and width of grid lines to make the plot more distinct.
```

<h2>Pair Grid Plots<h2>

```python
iris = sns.load_dataset('iris')
i1 = sns.PairGrid(iris,hue='species',palette='cool')
i1.map_diag(plt.hist)
i1.map_offdiag(plt.scatter)
plt.gcf().set_size_inches(15,8) 
```
