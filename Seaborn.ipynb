```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
df_crash = sns.load_dataset('car_crashes')
df_crash.head(3)
```

<h2>Distribution Plots<h2>
<h3>These include Distribution Plot, Joint Plot, KDE Plot, Pair Plot etc.<h3>

```python
#We'll first discuss distribution plot. It is used to plot a distribution of one variable.
sns.distplot(df_crash['not_distracted'])
```

```python
#Remove the Kernel Density Estimation(KDE) line.
sns.distplot(df_crash['not_distracted'],kde=False)
```

```python
#Alter the bin width.
sns.distplot(df_crash['not_distracted'],kde=False,bins=20)
```

```python
#We'll now look at Joint Plots. These are used to plot a distribution of two variables as a scatter plot.
sns.jointplot(x='speeding',y='alcohol',data=df_crash,kind='reg')
```

```python
sns.jointplot(x='speeding',y='alcohol',data=df_crash,kind='kde')
```

```python
sns.jointplot(x='speeding',y='alcohol',data=df_crash,kind='hex')
```

```python
#We'll now move to Pair Plots. It creates a matrix of plots showing relationship b/w all numeric variables.
#We'll first load 'tips' dataset.
df_tips = sns.load_dataset('tips')
df_tips.head(3)
```

```python
sns.pairplot(df_tips,hue='sex',palette='Blues')
plt.gcf().set_size_inches(15,8) 
#Alter the size of the output plot using 'plt.gcf()' method. Used AFTER the plot statement.
```

<h2>Styling<h2>

```python
sns.set_style('darkgrid') #Dark background with grids.
sns.jointplot(x='speeding',y='alcohol',data=df_crash,kind='reg')
```

```python
sns.set_style('white') #White background. We also have 'whitegrid', 'dark' and 'ticks' to name a few popular options.
sns.jointplot(x='speeding',y='alcohol',data=df_crash,kind='reg')
plt.gcf().set_size_inches(13,8)
```
