<h2>Bar Plots<h2>

```python
jsdev_y = [37321,43534,46456,49778,53124,56456,62487,66426,68210,68178,75125]
plt.bar(dev_x,dev_y,label='All Devs')
plt.bar(dev_x,pydev_y,label='Python Devs')
plt.bar(dev_x,jsdev_y,label='Java Devs')
plt.style.use('seaborn-poster')
plt.legend()
plt.show()
#As seen, the 3 bars are stacked over each other rather than be shown side-by-side.
#To resolve it, we include a width offset on each bar so that they appear separately rather than on top.
```

```python
import numpy as np
x_index = np.arange(len(dev_x))
width=0.25
plt.bar(x_index-width,dev_y,width=width,label='All Devs') #Shift the bar to the left.
plt.bar(x_index,pydev_y,width=width,label='Python Devs') #Bar remains as it is.
plt.bar(x_index+width,jsdev_y,width=width,label='Java Devs') #Shift the bar to the right.
plt.title('Media Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.legend()
plt.show()
#By including the width offset, the bars are now aligned side-by-side.
```

```python
#To offset our bars, we changed our x-axis from actual age values to indices. As offsetting actual data doesn't fit the pupose.
#However, it's not good to see indices instead of actual age figures on our bar plot.
#To include the ages again on x-axis, we use 'xticks' function as follows.
plt.xticks(ticks=x_index,labels=dev_x)
plt.bar(x_index-width,dev_y,width=width,label='All Devs') #Shift the bar to the left.
plt.bar(x_index,pydev_y,width=width,label='Python Devs') #Bar remains as it is.
plt.bar(x_index+width,jsdev_y,width=width,label='Java Devs') #Shift the bar to the right.
plt.title('Media Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.legend()
```

```python
from collections import Counter
c=Counter(['Auba','Laca','Saka','Auba','Saka'])
c #The 'Counter' function takes in a list argument and returns count of each value occuring in the list as key:value pairs.
```

```python
with open('D:/pandas Practice/data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file) #The 'DictReader' creates a dictionary object out of the 'data.csv' file.
    count_languages = Counter() #Create an empty counter which later updates and takes in list arguments from each iteration of 'for' loop.
    for i in csv_reader: #At the end of last iteration, it will return the count of each unique value of the list generated after split.
        count_languages.update(i['LanguagesWorkedWith'].split(';'))
print(count_languages)
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(count_languages,orient='index').reset_index()
df.rename(columns={'index':'LanguagesWorkedWith',0:'Count'},inplace=True)
df.head(3)
```

```python
#Now we have our original 'LanguagesWorkedWith' column as 2 separate columns, one having name of languages and other having the counts.
#Basically, we arrived at an object which tells us the count of each language as it appears in the original messy column.
#Now we can plot the languages with their popularity of usage.
plt.barh(df['LanguagesWorkedWith'],df['Count']) #Plot a horizontal bar instead of vertical.
plt.title('Usage of Languages')
plt.xlabel('Count of Users')
plt.ylabel('Languages')
```

<h2>Pie Charts<h2>

```python
slices = [15,25,35,25]
labels = ['Matlab','SQL','Python','R']
color = ['#d690bb','#a6edb9','#8890db','#deb78a'] #Hex color codes found online.
plt.pie(slices,labels=labels,colors=color)
plt.title('Analytical Tools by Popularity')
plt.show()
```

```python
#The 'explode' option takes in a list of values as a proportion. 
exp = [0,0,0.07,0] #Here we explode the 'Python' slice by 7% in the pie chart so that it pops out distinctly.
plt.pie(slices,labels=labels,colors=color,explode=exp,shadow=True,autopct='%1.1f%%',textprops={'fontsize': 14}) #'shadow' option includes a 3d effect on the chart.
#'textprops' option adjusts the size of labels on the plot.
plt.title('Proportion of Learners by Language',fontsize=22)
plt.show() #The 'autopct' option includes the percentage share of each slice in the pie-chart. The value '%1.1f%%' is given as it is.
```

<h2>Fill Area in Line Plot<h2>

```python
dev_x = [i for i in range(25,36)]
dev_y = [38496,42000,46752,49320,52300,56000,62316,64928,67317,68419,72148]
pydev_y = [45321,47234,49456,56378,60124,63456,65487,68426,70210,74178,75125]
plt.plot(dev_x,dev_y,label='All Devs')
plt.plot(dev_x,pydev_y,label='Python Devs')
plt.fill_between(dev_x,pydev_y,alpha=0.3) 
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.legend() #Option 'alpha' included to make the fill more transparent so that both lines are clearly visible.
plt.show()
```

```python
#Include a point where the fill shows differently above and below the point.
median_salary = 57_000
plt.plot(dev_x,dev_y,label='All Devs')
plt.plot(dev_x,pydev_y,label='Python Devs')
plt.fill_between(dev_x,pydev_y,median_salary,alpha=0.3) 
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.legend() 
plt.show()
```

```python
#We can use a 'where' option to indicate the color pattern based on a criteria.
plt.plot(dev_x,dev_y,label='All Devs')
plt.plot(dev_x,pydev_y,label='Python Devs')
plt.fill_between(dev_x,pydev_y,median_salary,where=([i>median_salary for i in pydev_y]),color='#0e8f5d',interpolate=True) 
plt.fill_between(dev_x,pydev_y,median_salary,where=([i<=median_salary for i in pydev_y]),color='#bd243d',interpolate=True) 
```

```python
#To fill between regions, specify first the x-axis argument(i.e.'dev_x')followed by the 2 y-axis arguments between which we want the fill.
jsdev_y = [37321,43534,43456,49778,53124,51456,62487,66426,68210,68178,75125]
plt.plot(dev_x,dev_y,label='All Devs')
plt.plot(dev_x,pydev_y,label='Python Devs')
plt.plot(dev_x,jsdev_y,label='Java Devs')
plt.fill_between(dev_x,jsdev_y,pydev_y,alpha=0.3) #Fill between all devs and java devs.
plt.fill_between(dev_x,dev_y,jsdev_y,alpha=0.3) #Fill between java devs and python devs.
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.legend() 
plt.show()
```

<h2>Scatter Plots<h2>

```python
d2 = pd.read_csv('D:/pandas Practice/data2.csv')
views = d2['view_count']
likes = d2['likes']
ratio = d2['ratio']
plt.scatter(views,likes,edgecolor='black',alpha=0.75,linewidth=1)
plt.title('Trending YT Videos')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
#As seen, an outlier data point distorts and disrupts the plot.
```

<h2>Plotting Time Series Data<h2>

```python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
from matplotlib import pyplot as plt 
import pandas as pd
```

```python
dates = [datetime(2021,10,24),datetime(2021,10,25),datetime(2021,10,26),datetime(2021,10,27),datetime(2021,10,28),datetime(2021,10,29),datetime(2021,10,30)]
y = [0,1,4,2,6,4,7]
```

```python
plt.plot_date(dates,y,linestyle='solid')
plt.gcf().autofmt_xdate() #Auto format the dates on x-axis so that they are aligned in a readable manner.
date_format = mpl_dates.DateFormatter('%d %b %Y') #Use 'DateFormatter' function to set a custom date format.
plt.gca().xaxis.set_major_formatter(date_format) #Supply the format to 'xaxis.set_major_formatter'.
plt.show()
```

```python
d5 = pd.read_csv('D:/pandas Practice/data3.csv')
date = d5['Date']
close = d5['Close']
plt.style.use('seaborn')
plt.plot_date(date,close,linestyle='solid')
plt.gcf().autofmt_xdate()
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
#As seen from plot, the dates exist as strings and not actual dates. 
#The last date 17-05-19 should be sorted as the first value if it was actual date type.
```

```python
date = pd.to_datetime(d5['Date']) #Convert date column from string to datetime object.
d5.sort_values('Date',inplace=True) #Sort the data on new date column and overwrite it.
plt.style.use('seaborn')
plt.plot_date(date,close,linestyle='solid')
plt.gcf().autofmt_xdate()
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
#Now we see, the added date 17-05-19 comes as first date value on x-axis.
```

<h2>Subplots<h2>
<h3>Used for creating a matrix of plots in one go.<h3>

```python
#We use 'subplots' function and specify the number of plots in matrix by giving number of rows and columns.
plt.subplots(nrows=2,ncols=2)
#As seen it gives a matrix of 2x2 for our plots.
#Also notice the output object returned is a figure and axes.
```

```python
d3 = pd.read_csv('D:/pandas Practice/data4.csv')
age = d3['Age']
all_devs = d3['All_Devs']
py_devs = d3['Python']
js_devs = d3['JavaScript']
plt.style.use('seaborn')
fig, ax = plt.subplots() #Call 'subplot' function and unpack its content as a figure and axes. 
ax.plot(age,py_devs,label='Python Devs')
ax.plot(age,js_devs,label='Java Devs')
ax.plot(age,all_devs,label='All Devs')
ax.set_title('Salary by Age of Different Devs')
ax.set_xlabel('Age')
ax.set_ylabel('Salaries')
ax.legend()
plt.show()
# A basic usage of subplots. No difference from the last plot, but we shall see as we go along.
```

```python
plt.style.use('seaborn')
fig ,(ax1,ax2) = plt.subplots(nrows=2,ncols=1,sharex=True) #Here we specified 2 rows and 1 column and unpacked the 2 axes as 'ax1' and 'ax2'.
ax1.plot(age,py_devs,label='Python Devs') #The 'sharex' option set to 'True' indicates both plots share the same x-axis i.e. no need to plot x-axis for both as seen.
ax1.plot(age,js_devs,label='Java Devs')
ax2.plot(age,all_devs,label='All Devs')
ax1.set_title('Salary by Age of Different Devs')
#ax1.set_xlabel('Age') #Not rquired as 'ax2' xlabel is sufficient.
ax1.set_ylabel('Salaries')
ax1.legend()
#ax2.set_title('Salary by Age of Different Devs') #Not required as 'ax1' title is sufficient.
ax2.set_xlabel('Age')
ax2.set_ylabel('Salaries')
ax2.legend()
plt.show()
```

```python
#Instead of 1 figure having 2 plots, we now plot 2 separate figures for each plot.
fig ,ax1 = plt.subplots()
fig ,ax2 = plt.subplots()
ax1.plot(age,py_devs,label='Python Devs') 
ax1.plot(age,js_devs,label='Java Devs')
ax2.plot(age,all_devs,label='All Devs')
ax1.set_title('Salary by Age of Different Devs')
ax1.set_xlabel('Age') 
ax1.set_ylabel('Salaries')
ax1.legend()
ax2.set_title('Salary by Age of Different Devs') 
ax2.set_xlabel('Age')
ax2.set_ylabel('Salaries')
ax2.legend()
plt.show()
```
