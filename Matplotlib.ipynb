<h2>Line Plots<h2>

```python
dev_x = [i for i in range(25,36)]
dev_y = [38496,42000,46752,49320,52300,56000,62316,64928,67317,68419,72148]
plt.plot(dev_x,dev_y) #Use 'plot' function to create a simple line plot between 2 continuous variables.
plt.show()
```

```python
#Include labels for both axes and plot title.
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.title('Median Salary(USD) by Age')
plt.plot(dev_x,dev_y)
plt.show()
```

```python
#For a common x-axis, we plot a line chart for 2 different y-axis variables.
pydev_y = [45321,47234,49456,56378,60124,63456,65487,68426,70210,74178,75125]
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.title('Median Salary(USD) by Age')
plt.plot(dev_x,dev_y,label='For all devs') #Included 2 plot functions, one for each line chart.
plt.plot(dev_x,pydev_y,label='For Python devs') 
plt.legend() #To differentiate b/w the lines, we include 'label' option in both 'plot' functions and then include a legend.
plt.show()
```

```python
#We do some color, linestyle and marker related customization to our line plot.
plt.plot(dev_x,dev_y,color='#444444',linestyle='--',marker='o',linewidth=3,label='For all devs') 
plt.plot(dev_x,pydev_y,color='#5a79da',marker='.',linewidth=3, label='For Python devs') 
plt.legend() #The values in 'color' option are hex color values. Can be taken online as per need.
plt.grid(True)
```

<h2>Bar Plots<h2>

```python
jsdev_y = [37321,43534,46456,49778,53124,56456,62487,66426,68210,68178,75125]
plt.bar(dev_x,dev_y,label='All Devs')
plt.bar(dev_x,pydev_y,label='Python Devs')
plt.bar(dev_x,jsdev_y,label='Java Devs')
plt.style.use('seaborn-poster')
plt.legend()
plt.show()
#As seen, the 3 bars are stacked over each other rather than be shown side-by-side.
#To resolve it, we include a width offset on each bar so that they appear separately rather than on top.
```

```python
import numpy as np
x_index = np.arange(len(dev_x))
width=0.25
plt.bar(x_index-width,dev_y,width=width,label='All Devs') #Shift the bar to the left.
plt.bar(x_index,pydev_y,width=width,label='Python Devs') #Bar remains as it is.
plt.bar(x_index+width,jsdev_y,width=width,label='Java Devs') #Shift the bar to the right.
plt.title('Media Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.legend()
plt.show()
#By including the width offset, the bars are now aligned side-by-side.
```

```python
#To offset our bars, we changed our x-axis from actual age values to indices. As offsetting actual data doesn't fit the pupose.
#However, it's not good to see indices instead of actual age figures on our bar plot.
#To include the ages again on x-axis, we use 'xticks' function as follows.
plt.xticks(ticks=x_index,labels=dev_x)
plt.bar(x_index-width,dev_y,width=width,label='All Devs') #Shift the bar to the left.
plt.bar(x_index,pydev_y,width=width,label='Python Devs') #Bar remains as it is.
plt.bar(x_index+width,jsdev_y,width=width,label='Java Devs') #Shift the bar to the right.
plt.title('Media Salary(USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary(USD)')
plt.legend()
```

```python
from collections import Counter
c=Counter(['Auba','Laca','Saka','Auba','Saka'])
c #The 'Counter' function takes in a list argument and returns count of each value occuring in the list as key:value pairs.
```

```python
with open('D:/pandas Practice/data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file) #The 'DictReader' creates a dictionary object out of the 'data.csv' file.
    count_languages = Counter() #Create an empty counter which later updates and takes in list arguments from each iteration of 'for' loop.
    for i in csv_reader: #At the end of last iteration, it will return the count of each unique value of the list generated after split.
        count_languages.update(i['LanguagesWorkedWith'].split(';'))
print(count_languages)
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(count_languages,orient='index').reset_index()
df.rename(columns={'index':'LanguagesWorkedWith',0:'Count'},inplace=True)
df.head(3)
```

```python
#Now we have our original 'LanguagesWorkedWith' column as 2 separate columns, one having name of languages and other having the counts.
#Basically, we arrived at an object which tells us the count of each language as it appears in the original messy column.
#Now we can plot the languages with their popularity of usage.
plt.barh(df['LanguagesWorkedWith'],df['Count']) #Plot a horizontal bar instead of vertical.
plt.title('Usage of Languages')
plt.xlabel('Count of Users')
plt.ylabel('Languages')
```

```python
#We can alternately use 'pandas' instead of 'open' and 'DictReader'.
data = pd.read_csv('D:/pandas Practice/data.csv')
ids = data['Responder_id']
languages_known = data['LanguagesWorkedWith']
lang_counter = Counter()
for i in languages_known:
    lang_counter.update(i.split(';'))
print(lang_counter)
```

<h2>Pie Charts<h2>

```python
slices = [15,25,35,25]
labels = ['Matlab','SQL','Python','R']
color = ['#d690bb','#a6edb9','#8890db','#deb78a'] #Hex color codes found online.
plt.pie(slices,labels=labels,colors=color)
plt.title('Analytical Tools by Popularity')
plt.show()
```

```python
#The 'explode' option takes in a list of values as a proportion. 
exp = [0,0,0.07,0] #Here we explode the 'Python' slice by 7% in the pie chart so that it pops out distinctly.
plt.pie(slices,labels=labels,colors=color,explode=exp,shadow=True,autopct='%1.1f%%') #'shadow' option includes a 3d effect on the chart.
plt.show() #The 'autopct' option includes the percentage share of each slice in the pie-chart. The value '%1.1f%%' is given as it is.
```

<h2>Area Plots<h2>
<h3>Used mainly when we have timestamps on x-axis and we want to plot productivity/output of multiple individuals/entities on y-axis.

```python
players = ['Auba','ESR','Saka']
timestamps = [15,30,45,60,75,90]
auba = [2.2,3.5,4.2,5.7,7.5,8.3] #Ground covered in kms. till given timestamps.
esr = [2.5,4.1,5.3,6.8,7.9,9.5]
saka = [2.4,3.6,4.2,5.5,6.7,7.9]
color = ['#d690bb','#a6edb9','#8890db']
plt.stackplot(timestamps,auba,esr,saka,labels=players,colors=color)
plt.legend(loc='upper left') #Option to customize the location of legend as per the plot.
plt.show()
```

```python
days = ['Mon','Tue','Wed','Thu','Fri']
Edward = [5,6,4,6,5]
Hahn = [3,4,2,5,3]
Chang = [2,4,3,6,5]
devs = ['Edward','Hahn','Chang']
plt.stackplot(days,Edward,Hahn,Chang,labels=devs)
plt.ylabel('Work Hrs.')
plt.legend(loc='upper left')
plt.show()
```
